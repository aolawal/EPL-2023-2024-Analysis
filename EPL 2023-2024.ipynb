## Converting all the columns into the appropriate datatype

```python
df_column = ['Team', 'Form, Last 6 games, Oldest first']
for column in df_column:
    df_epl[column] = df_epl[column].astype("string")

df_epl.info()
```

## Seperating the Text in Team Column from the Team position number

```python
df_epl[['Position', 'Team']] = df_epl['Team'].str.extract(r'(\d+)([^\d]+)')
```

## Statitsics of the Top 5 Teams in EPL

```python
Top5_Statistics= df_epl[['Team', 'Won', 'Drawn', 'Lost','Points']].nlargest(5, 'Points')

Top5_Statistics

Top5_Statistics_excl = Top5_Statistics[['Team', 'Won', 'Drawn', 'Lost']]

Top5_Statistics_excl
```

## Performance of the Top 5 Teams on EPL table

```python
import matplotlib.pyplot as plt

# Define custom colors for Win, Draw, Loss — in the correct column order
custom_colors = ['green', 'skyblue', 'orange']

# Plot the stacked bar chart
ax = Top5_Statistics_excl.plot(
    kind='bar',
    x='Team',
    stacked=True,
    color=custom_colors,
    title='Performance of the Top 5 teams in the EPL',
    xlabel='Team',
    ylabel='Played',
    figsize=(10, 6)
)

# Add data labels
for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        if height > 0:
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f'{int(height)}',
                ha='center',
                va='center',
                color='white',
                fontsize=9
            )

plt.tight_layout()
plt.show()
```
